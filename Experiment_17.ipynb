```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
np.random.seed(0)
df=pd.DataFrame({
    'Category':['A','B','C','D','E'],
    'Value':[60,20,10,40,90],
    'Sales':np.random.randint(100,500,5),
    'Profit':np.random.randint(10,100,5)

})
print(df.head())
```

```python
# 1. AREA PLOT
plt.figure()
plt.fill_between(df['Category'],df['Value'],color='pink',alpha=0.4)# APLHA IS FOR INTENSITY
plt.title('Area Plot')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# PIE CHART
plt.figure()
plt.pie(df['Value'],labels=df['Category'],autopct='%0.2f%%')
plt.title('Pie chart ')
plt.show()
```

```python
from numpy.random.mtrand import f
# DONUT CHART
plt.figure()
plt.pie(df['Value'],labels=df['Category'],autopct='%1.1f%%')
center_circle =plt.Circle((0,0),0.5,fc='white')
fig=plt.gcf()
fig.gca().add_artist(center_circle)
plt.title('Donut chart')
plt.show()
```

```python
# BOXPLOT (OUTLIERS DETECTION)
plt.figure()
sns.boxplot(x=df['Value'],color='skyblue')
plt.title('Boxplot of sales')
plt.show()
```

```python
# HEATMAP (CORRELATION)
plt.figure()
corr =df[['Sales','Profit']].corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title("Correlation heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt

# BUBBLE PLOT
plt.figure()
plt.scatter(df['Sales'],df['Profit'],
            s=df['Value']*100,
            c=df['Value'],
            cmap='magma')
plt.ylabel('Profit')
plt.title('Bubble Plot')
plt.show()
```

```python
#
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,4))

sns.scatterplot(x='Sales',y='Profit',
                size='Value',
                hue='Value',
                data=df,
                sizes=(100,1000),
                palette='coolwarm')
plt.title("Bubble plot using seaborn")
plt.show()
```

```python
# SECOND DATASET
import numpy as np
import pandas as pd
np.random.seed(7)
n = 100

df = pd.DataFrame({
    'Customer_ID' : range(1, n+1),
    'Age' : np.random.randint(20,60,n),
    'Income' : np.random.randint(20000,100000,n),
    'Loan_Amount' : np.random.randint(5000, 50000, n),
    'Credit_Score' : np.random.randint(300,850,n)
})

df
```

```python
# 1. AREA PLOT
import matplotlib.pyplot as plt


plt.subplot(3, 2, 1)
df_sorted = df.sort_values('Age').reset_index()
plt.fill_between(df_sorted.index, df_sorted['Income'], label='Income', alpha=0.5, color='skyblue')
plt.fill_between(df_sorted.index, df_sorted['Loan_Amount'], label='Loan Amount', alpha=0.5, color='orange')
plt.title(' Area Plot: Income vs Loan Amount')
plt.xlabel('Index (Sorted by Age)')
plt.ylabel('Amount ($)')
plt.legend()
```

```python
# 2. PIE CHART

plt.figure()
plt.pie(df['Income'], labels = df['Credit_Score'], radius = 4)
plt.title("Pie Chart")
plt.show()
```

```python
# 3. BOX PLOT (OUTLIER DETECTION)
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['Income', 'Loan_Amount']])
plt.title('Distribution of Income and Loan Amount')
```

```python
# 4. HEATMAP

plt.figure(figsize=(8, 6))
sns.heatmap(df[['Age', 'Income', 'Loan_Amount', 'Credit_Score']].corr(), annot=True, cmap='magma')
plt.title('Correlation Heatmap')
```

```python
# 5. BUBBLE PLOT

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Income'], s=df['Loan_Amount']/100, alpha=0.5, c=df['Credit_Score'], cmap='viridis')
plt.colorbar(label='Credit Score')
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Age vs Income (Size=Loan, Color=Credit Score)')
```
